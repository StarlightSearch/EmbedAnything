```python
import embed_anything
import numpy as np
from PIL import Image
```

```python
data = embed_anything.embed_directory("D:\Projects\zero-shot\images", embeder= "Clip")
embeddings = np.array([data.embedding for data in data])
print(data[0].embedding)
```

```python
for i in range(len(data)):
    print(data[i], data[i].embedding)
```

```python
labels = ['Shirt', 'Coat', 'Trousers', 'Skirt', 'Hat', 'Shoes', 'Bag']
```

```python
label_embeddings = []
for i in labels:
    text = embed_anything.embed_query([i], embeder= "Clip")
    print(text[0].embedding)
    label_embeddings.append(text[0].embedding)
```

```python
import matplotlib.pyplot as plt
for i in range(len(data)):
    similarities = np.dot(label_embeddings, data[i].embedding)
    max_index = np.argmax(similarities)
    
    image_path = data[i].text
    
    # Open and plot the image
    img = Image.open(image_path)
    plt.imshow(img)
    plt.title(labels[max_index])
    plt.axis('off')
    plt.show()
```
